```python
import sqlite3 as sql
import pandas as pd
import numpy as np
```

```python
CONN = sql.connect("/Users/divya/projects/pandas/divvy-example/Divvy_Trips.db")
stations = "SELECT * FROM temp WHERE [FROM STATION ID] = 90 OR [TO STATION ID] = 90";
df = pd.read_sql(stations, con = CONN)
CONN.close()
```
